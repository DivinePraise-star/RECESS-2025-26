```python
# Housing Data Analysis
#Absolute maximum scaling of housing data
```

```python
#libraries
import pandas as pd
import numpy as np

df = pd.read_csv('housing.csv')
df.head()
```

```python
df = df.select_dtypes(include=[np.number])  # Select only numeric columns
df.head(5)   
```

```python

#maxscaling
max_abs = df.abs().max()  # Get the absolute maximum values for each column
#or
max_abs = np.max(np.abs(df), axis=0)  # Get the absolute maximum values for each column using NumPy
scaled_df = df / max_abs  # Scale the DataFrame by dividing by the absolute maximum values
scaled_df.head(5)  # Display the first 5 rows of the scaled DataFrame
```
