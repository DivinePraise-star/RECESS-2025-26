```python
#line graph
#it is a type of chart that displays information as a series of data points connected by straight line segments.    
#they are used to display data that changes over time, such as stock prices or temperature readings.    
#they are plotted using points connected by lines, where the x-axis represents time and the y-axis represents the value of the data being displayed.

#libraries
import matplotlib.pyplot as plt
```

```python
#create data
x = [1, 2, 3, 4, 5, 6]
y = [2, 4, 6, 8, 10, 12]
```

```python
#plot data
fig, ax = plt.subplots() #fig is the figure object, which contains the entire figure, including axes, labels, and other elements
ax.plot(x, y) #ax is the axes object, which allows for more customization and control over the plot
#or
plt.plot(x, y)

#the plot dots on the graph can be customized by adding a marker argument to the plot function. For example, plt.plot(x, y, marker='o') will add circular markers to each data point on the line graph. 
plt.plot(x, y, marker='o')

#customize plot
plt.title("A Line Graph Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

#key with legend and label

plt.legend(['Data'])

#Display the plot
plt.show()
```

```python
#Bar chart
#it is a type of chart that presents categorical data with rectangular bars, where the length of each bar is proportional to the value it represents.   
#used to compare different categories of data. 
# It is useful for visualizing discrete data and showing the differences between groups.   
#they are often used to display data that is organized into categories, such as sales figures for different products or the number of students in different classes.    
#they are plotted using rectangular bars, where the height of each bar represents the value of the data being displayed.

#libraries
import matplotlib.pyplot as plt
```

```python
#create data
x = ['Kampala', 'Nairobi', 'Dar es Salaam', 'Kigali', 'Bujumbura', 'Kigali', 'Kampala', 'Nairobi', 'Dar es Salaam', 'Kigali']
y = [37, 45, 28.5, 23.6, 21.5, 23.6, 37, 45, 28.5, 23.6]
```

```python
#plot data
plt.bar(x, y)

#customize plot with color, width, and edge color and labels
#plt.bar(x, y, color='blue', width=0.5, edgecolor='black')
plt.title("A bar chart Example")
plt.xlabel("Cities")
plt.ylabel("Temperature in Celsius degrees")

#key with legend and label
plt.legend(['Temperature'])

#mark the highest and lowest points on the bar chart with a different color and label
#plt.bar(x[y.index(max(y))], max(y), color='red', label='Highest Point')
#plt.bar(x[y.index(min(y))], min(y), color='green', label='Lowest Point')

#Display the plot
plt.show()
```

```python
#Scatter plot
#it is a type of chart that displays data as a collection of points, where each point represents a value in a two-dimensional space.
#they are used to visualize the relationship between two variables, and can help identify patterns, trends, and outliers in the data.
#they are plotted using points on a Cartesian plane, where the x-axis represents one variable and the y-axis represents the other variable. 

#libraries
import matplotlib.pyplot as plt
```

```python
#create data
x = ['A', 'B', 'C', 'D', 'E']
y = [23, 45, 56, 78, 32]
```

```python
#plot data
plt.plot(x, y)

#customize plot with color, marker, and linestyle
plt.scatter(x, y, color='blue', marker='o', linestyle='--') 
plt.title("A Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

#legend with label
plt.legend(['Data'])

#display the plot
plt.show()
```

```python
#pie chart
#it is a type of chart that displays data as a circle divided into slices, where each slice represents a proportion of the whole.
#they are used to visualize the relative proportions of different categories of data, and can help identify patterns and trends in the data.    
#they are plotted using a circular shape, where each slice represents a category of data and the size of each slice is proportional to the value of that category.  

#libraries
import matplotlib.pyplot as plt
```

```python
#create data
x = ['youth', 'adults', 'elderly']
y = [55, 25, 20]    
```

```python
#plot data
plt.pie(y, labels = x)
plt.legend(['youth', 'adults', 'elderly'])
plt.title("A Pie Chart Example")
plt.show()
```

```python
#box plot
#it is a type of chart that displays the distribution of a dataset by showing the median, quartiles, and outliers of the data.      
#they are used to visualize the spread and skewness of a dataset, and can help identify patterns and trends in the data.    
#they are plotted using a rectangular box, where the box represents the interquartile range (IQR) of the data, and the whiskers represent the minimum and maximum values of the data. Outliers are plotted as individual points outside of the whiskers.    

#libraries
import matplotlib.pyplot as plt
```

```python
#create data
x = ['A', 'B', 'C', 'D', 'E']
y = [25, 45, 30, 35, 20]
```

```python
#plot data
plt.boxplot(y)
plt.xticks(range(1, len(x) + 1), x)
plt.title("A Box Plot Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.legend(['Data'])
plt.show()
```

```python
#histogram
#it is a type of chart that displays the distribution of a dataset by dividing the data into intervals, or bins, and counting the number of data points that fall into each bin.    
#they are used to visualize the frequency distribution of a dataset, and can help identify patterns and trends in the data. 
#they are plotted using rectangular bars, where the height of each bar represents the number of data points that fall into each bin.    

import matplotlib.pyplot as plt
```

```python
#create data
x = ['A', 'B', 'C', 'D', 'E']
y = [5, 17, 24, 16, 8]
```

```python
#plot data
plt.hist(y, bins=5, color='lightblue', edgecolor='black')
plt.title("A Histogram Example")
plt.xlabel("Grades")
plt.ylabel("Number of Students")
plt.show()
```
