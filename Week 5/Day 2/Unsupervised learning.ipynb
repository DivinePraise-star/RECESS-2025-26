```python
#Unsupervised learning 
# is a type of machine learning where the model is trained on unlabeled data, meaning that the input data does not have corresponding output labels. 
# The goal of unsupervised learning is to find patterns, structures, or relationships within the data without any prior knowledge of what the output should be. 
#data has no labels, and the algorithm tries to identify patterns or groupings in the data.

#Real-world examples of unsupervised learning include:
#1. Customer Segmentation: Businesses can use unsupervised learning to segment their customers into different groups based on purchasing behavior, demographics, or other characteristics. This can help in targeted marketing and personalized recommendations.
#2. Anomaly Detection: Unsupervised learning can be used to identify unusual patterns or anomalies in data, such as detecting fraudulent transactions in financial data or identifying defects in manufacturing processes.
#3. Market Basket Analysis: Retailers can use unsupervised learning to analyze customer purchase patterns and identify associations between products. This can help in optimizing product placement and cross-selling strategies.
#Image and Video Analysis: Unsupervised learning can be applied to image and video data to discover patterns, such as clustering similar images or identifying objects in videos without explicit labels.
#Document Clustering: Unsupervised learning can be used to group similar documents together based on their content, which can be useful for organizing large collections of text data, such as news articles or research papers.
#Social Network Analysis: Unsupervised learning can be applied to social network data to identify communities or clusters of users with similar interests or connections.
#Dimensionality Reduction: Unsupervised learning techniques like Principal Component Analysis (PCA) can be used to reduce the dimensionality of data while preserving important information, which can be useful for visualization and feature extraction.
#Genetic Data Analysis: Unsupervised learning can be used to analyze genetic data and identify patterns or clusters of genes that may be associated with certain traits or diseases.
```

```python
#Example1
#Customer Segmentation using K-Means Clustering:
#A supermarket wants to segment its customers based on their purchasing behavior. They have collected data on various features such as the total amount spent, frequency of visits, and types of products purchased. The supermarket can use K-Means clustering, an unsupervised learning algorithm, to group customers into distinct segments based on these features. Each cluster may represent a different type of customer, such as high-spending frequent shoppers or occasional bargain hunters. This segmentation can help the supermarket tailor marketing strategies and promotions to each customer group.   
#Customer, age, monthly income, spending score, and frequency of visits are the features used for clustering. The K-Means algorithm will assign customers to clusters based on their similarity in these features, allowing the supermarket to identify distinct customer segments and develop targeted marketing strategies for each group.    

#Fraud detection using Isolation Forest:
#A bank wants to detect fraudulent transactions in its credit card data. They have a dataset of transactions, but they do not have labeled examples of which transactions are fraudulent. The bank can use the Isolation Forest algorithm, an unsupervised learning technique, to identify anomalies in the transaction data. The algorithm will isolate transactions that deviate significantly from the normal patterns of behavior, flagging them as potential fraud. This allows the bank to investigate these flagged transactions further and take appropriate action to prevent financial losses.
#Transaction ID, transaction amount, time of transaction, and location are the features used for anomaly detection. The Isolation Forest algorithm will analyze these features to identify transactions that are significantly different from the majority of transactions, allowing the bank to detect potential fraudulent activity without needing labeled examples of fraud.    
#Normal transactions will cluster together, while fraudulent transactions will be isolated and flagged for further investigation.   
```

```python
#Example 2
#Image Clustering using K-Means in unsupervised learning:
#A company wants to organize a large collection of images into meaningful groups based on visual similarity. 
# They can use K-Means clustering, an unsupervised learning algorithm, to group similar images together. The algorithm will analyze the pixel values and features of the images to identify patterns and create clusters of visually similar images.
#For example, the company may have a dataset of images containing different types of animals.

#What is clustering in unsupervised learning?
#Clustering is a technique used in unsupervised learning to group similar data points together based on their features or characteristics. The goal of clustering is to identify natural groupings or patterns in the data without any prior knowledge of the labels or categories. Clustering algorithms analyze the similarities and differences between data points and assign them to clusters based on their proximity in the feature space.

#What is K-Means clustering in unsupervised learning?
#K-Means clustering is a popular unsupervised learning algorithm used for clustering data points into K distinct clusters. The algorithm works by iteratively assigning data points to the nearest cluster centroid and updating the centroids based on the mean of the assigned points. The process continues until convergence, where the assignments no longer change significantly. K-Means is widely used for its simplicity and efficiency in clustering large datasets. 
```

```python
#libraries for unsupervised learning
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

```python
# ----------------------------------------------------------
# Image Names
# (Used only for displaying results)
# ----------------------------------------------------------

image_names = [

    "dog_1.png","dog_2.png","dog_3.png",
    "dog_4.png","dog_5.png","dog_6.png",

    "cat_1.png","cat_2.png","cat_3.png",
    "cat_4.png","cat_5.png","cat_6.png",

    "car_1.png","car_2.png","car_3.png",
    "car_4.png","car_5.png","car_6.png",

    "mountain_1.png","mountain_2.png","mountain_3.png",
    "mountain_4.png","mountain_5.png","mountain_6.png",

    "unknown_1.png",
    "unknown_2.png",
    "unknown_3.png",
    "unknown_4.png",
    "unknown_5.png",
    "unknown_6.png"

]
```

```python
# ----------------------------------------------------------
# Sample Image Features
#
# Suppose every image is converted into two numerical
# features:
#
# Feature 1 = Color
# Feature 2 = Texture
#
# These values are made up ONLY for demonstration.
# ----------------------------------------------------------
#Our algorithm will not know if the images are dogs, cats, cars, or mountains. It will only see the numerical features and try to group them based on similarity.   

image_features = np.array([

# Dogs
[12,18],[14,20],[13,17],[15,19],[11,18],[16,20],

# Cats
[60,65],[62,67],[59,64],[63,66],[61,63],[60,68],

# Cars
[120,20],[118,18],[122,22],[119,19],[121,21],[123,20],

# Mountains
[200,180],[198,182],[201,179],
[203,181],[199,178],[202,183],

# Unknown Images
[250,50],     # Flower
[180,20],     # Airplane
[170,100],    # Bicycle
[90,150],     # City
[230,90],     # Burger
[180,200]     # Beach

])
```

```python
# Create KMeans Model
# ----------------------------------------------------------
kmeans = KMeans(
n_clusters=4,
random_state=42
)

# Train the model
kmeans.fit(image_features)

# Cluster assignments
clusters = kmeans.labels_
```

```python
# Display Results
# ----------------------------------------------------------

print("="*50)
print("IMAGE CLUSTERING RESULTS")
print("="*50)

for image, cluster in zip(image_names, clusters):

    print(f"{image:18} ---> Cluster {cluster}")

# ----------------------------------------------------------
# Cluster Centers
# ----------------------------------------------------------

print("\nCluster Centers")
print("-"*40)

print(kmeans.cluster_centers_)
```
